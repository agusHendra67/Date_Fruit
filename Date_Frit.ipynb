```python
##DATA PREPROCESSING
```

```python
##Skala terlalu kecil harus dikalikan dengan 100
small = ['SHAPEFACTOR_1','SHAPEFACTOR_2']
```

```python
for col in small:
    df[col] = df[col]*100
```

```python
from sklearn.preprocessing import LabelEncoder

ord = LabelEncoder()
data = ord.fit_transform(df['Class'])  ##INGAT HARUS BISA DI INVERSE LAGI KE NILAI AWALNYA
df['Class'] = data
```

```python
df['Class'].value_counts()
```

```python
corr = df.corr()

plt.figure(figsize=(36,72))
g = sns.heatmap(corr, vmax=.3, center=0, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt='.2f', cmap='coolwarm', square=True)
sns.despine()
plt.show()
```
